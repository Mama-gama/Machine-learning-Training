```python
import matplotlib.pyplot as plt
x=db.Gross_Worldwide
y=db.Rate
plt.plot(x,y,'g--')
plt.xlabel('Gross Worldwide')
plt.ylabel('Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=db.Rate
y=db.Metascore
plt.scatter(x,y,c='grey',marker='o')
plt.xlabel('IMDB Score')
plt.ylabel('Metascore')
plt.title('ratings')
plt.show()
```
